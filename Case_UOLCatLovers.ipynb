````python
#Exercício 1

import requests
import pandas as pd

def consultar_api():
    url = "https://cat-fact.herokuapp.com/facts"
    resposta = requests.get(url)
    if resposta.status_code == 200:
        return resposta.json()
    else:
        print("Falha ao consultar a API.")
        return None

def salvar_csv(dados):
    if dados:
        df = pd.DataFrame(dados)
        df.to_csv('cat_facts.csv', index=False)

def main():
    dados = consultar_api()
    salvar_csv(dados)

main()
````

Exercício 3


```
CREATE TABLE catlovers.cat_facts
(
  status STRING,
  _id STRING,
  user STRING,
  text STRING,
  __v INT64,
  source STRING,
  updatedAt TIMESTAMP,
  type STRING,
  createdAt TIMESTAMP,
  deleted BOOLEAN,
  used BOOLEAN
)
PARTITION BY TIMESTAMP_TRUNC(updatedAt, MONTH)
CLUSTER BY user, text, type, deleted;
```



Exercício 4



```
SELECT
  status,
  _id,
  user,
  text,
  __v,
  source,
  updatedAt,
  type,
  createdAt,
  deleted,
  used
FROM catlovers.cat_facts
WHERE EXTRACT(YEAR FROM updatedAt) = 2020
  AND EXTRACT(MONTH FROM updatedAt) = 8;
```
